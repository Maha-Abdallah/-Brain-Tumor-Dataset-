```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Load the dataset
df = pd.read_csv("brain_tumor_dataset.csv")
```

```python
# Data Cleaning
# Drop any rows with missing values
df.dropna(inplace=True)
```

```python
# Encode categorical columns (Tumor Type and Gender)
le = LabelEncoder()
df['Tumor Type'] = le.fit_transform(df['Tumor Type'])
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
# Data Visualization
# Example: Distribution of Tumor Types
plt.figure(figsize=(8, 6))
sns.countplot(x='Tumor Type', data=df)
plt.title("Distribution of Tumor Types")
plt.xlabel("Tumor Type")
plt.ylabel("Count")
plt.show()
```

```python
# Plot average tumor size by tumor type
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Tumor Type', y='Size (cm)', palette='viridis')
plt.title('Tumor Size by Tumor Type')
plt.xlabel('Tumor Type')
plt.ylabel('Size (cm)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Example: Age Distribution 
plt.figure(figsize=(8, 6))
sns.histplot(df['Patient Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Patient Age")
plt.ylabel("Count")
plt.show()
```

```python
# Plot gender distribution by tumor location
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Location', hue='Gender', palette='viridis')
plt.title('Gender Distribution by Tumor Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
